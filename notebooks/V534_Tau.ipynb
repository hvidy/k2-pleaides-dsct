# V534 Tau

```python
import lightkurve as lk
```

```python
target = 'V534 Tau'

search = lk.search_lightcurve(target, cadence='short', mission='K2')
search
```

```python
og_lc = search.download()
```

```python
og_lc.scatter();
```

```python
og_lc.remove_outliers().scatter();
```

```python
ax = og_lc.remove_outliers().scatter()
ax.set_xlim(2230,2235);
```

Apart from a few outliers, this light curve has no major problems, so there is no need to make a new one.
