# V650 Tau

```python
import lightkurve as lk
```

Let's take a look at the original light curve from the Kepler pipeline.

```python
target = 'V650 Tau'

search = lk.search_lightcurve(target, cadence='short', mission='K2')
search
```

```python
og_lc = search.download()
```

```python
og_lc.scatter();
```

```python
ax = og_lc.scatter()
ax.set_xlim(2230,2235);
```

This light curve is perfectly cromulent. There is no need to make a new one.
